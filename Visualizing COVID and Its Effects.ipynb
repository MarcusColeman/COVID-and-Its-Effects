# Problem Statement

### To Understand COVID's Inherent Trends, Patterns and Relationships as well as Associate Developments of the Pandemic to Various Aspects of Society

# Research Questions

## Research questions encompass two phases. The first phase's aim is to explore the dynamics of COVID. The second phase's goal is identify the impact the virus has had on society.

## Phase I

What trend(s) exists in New Cases, Total Cases, Total Deaths, Vaccinations, and Testing?

Is testing in the US an indicator of the change in total deaths worldwide?

Does a relationship exists between testing in the US and new cases?

Are total deaths, total cases, and new cases correlated across countries?

What's the relationship between US deaths and deaths globally?

Does a correlation exist between US total deaths and new cases world wide?

Does a correlation exist between deaths world wide and cases in the US?

What is the relationship between cases and deaths in the US?

What effect does new cases ahve on the change in testing and total deaths in the US?

How has vaccinations varied over time globally?

What effect has US and Global vaccinations had on total cases?

To what extext has US vaccinations altered the number of deaths and total cases?

How has vaccinations globally changed the number of reported cases and deaths?

## Phase II

To what extent has ongoing developments regarding the pandemic influenced activity in the US stock market


How has COVID impacted changes in employment in the US

### Testing

```python
testing_data=pd.read_csv('covid-testing-all-observations.csv')
```

```python
testing_data.head()
```

```python
#a dict to hold each country's testing dataframe
countries_testing={}

#creating list of unique countries
countries_testing_list=[country for country in np.unique(testing_data['ISO code'])]

#populating countries_testing dict
for country in countries_testing_list:
    #add key value (i.e. dataframe) to countries dict
    countries_testing[country]=testing_data[testing_data['ISO code']==country]
```

```python
#getting date ranges for testing data
testing_date_ranges={}

#iterating over countries_testing_list
for country in countries_testing_list:
    #assign key value pair
    testing_date_ranges[country]=(countries_testing[country].iloc[0]['Date'],countries_testing[country].iloc[-1]['Date'])
```

```python
def get_testing_data_range(country_symbol):
    return testing_date_ranges[country_symbol]
```

```python
def get_covid_closes(spy,covid_dataset):
    
    #get covid dataset dates
    dates=covid_dataset['date']
    
    #list to hold closes
    closes=[]
    
    for date in dates:
        if date +" 00:00:00" in list(spy_close.index):
            close=spy_close.loc[date +" 00:00:00"]['Close']
            closes.append(close)
        
    covid_dataset['Close']=closes
    
    return covid_dataset
```
